```python
def add_average_time_column(db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    try:
        cursor.execute("""
            ALTER TABLE CampaignPerformance
            ADD COLUMN average_time_in_budget REAL;
        """)
        conn.commit()
        print("Column average_time_in_budget added successfully.")
    except sqlite3.Error as e:
        # Column probably already exists
        print(f"SQLite info: {e}")
    finally:
        conn.close()
add_average_time_column("ads_data.db")
```

```python
def update_average_time_in_budget(db_path, df):
    """
    Updates CampaignPerformance.average_time_in_budget
    using values from a pandas DataFrame.
    """
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    try:
        for _, row in df.iterrows():
            cursor.execute("""
                UPDATE CampaignPerformance
                SET average_time_in_budget = ?
                WHERE campaign_name = ?
            """, (
                row['average_time_in_budget'],
                row['campaign_name']
            ))

        conn.commit()
        print(f"Updated average_time_in_budget for {cursor.rowcount} rows.")

    except sqlite3.Error as e:
        print(f"Database error: {e}")

    finally:
        conn.close()
update_average_time_in_budget('ads_data.db', df)
```
